## Step 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

#### **NOTE** I already completed this during the exercise and exported the data set, which is what I imported at the beginning of this task. So, I copied the code I used here but did not run any of it, except the checks.

```python
# Grouping "user_id" column, applying transform() on "orders_number" column to get max orders, and creating new column "max_order"
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Creating "Loyal customer" condition based off of "max_order" column
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Creating "Regular customer" condition based off of "max_order" column
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Creating "New customer" condition based off of "max_order" column
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking frequency of "loyalty_flag" column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## Step 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

    If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
    If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Grouping "user_id" column, applying transform() on "prices" column to get average amount spent, and creating new column "avg_spent"
ords_prods_merge['avg_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking dataframe for new "avg_spent" column
ords_prods_merge.head()
```

```python
# Creating "Low spender" condition based off of "avg_spent" column
ords_prods_merge.loc[ords_prods_merge['avg_spent'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Creating "High spender" condition based off of "avg_spent" column
ords_prods_merge.loc[ords_prods_merge['avg_spent'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking frequency of "spending_flag" column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Checking dataframe for new "spending_flag" column
ords_prods_merge.head()
```

## Step 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

    If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
    If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Grouping "user_id" column, applying transform() on "days_since_prior_order" column to get median amount of days since prior order, and creating new column "median_order_frequency"
ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking dataframe for new "median_order_frequency" column
ords_prods_merge.head()
```

```python
# Creating "Non-frequent customer" condition based off of "median_order_frequency" column
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating "Regular customer" condition based off of "median_order_frequency" column
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] > 10) & (ords_prods_merge['median_order_frequency'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Creating "Frequent customer" condition based off of "median_order_frequency" column
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Checking frequency of "order_frequency_flag" column
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Checking dataframe for new "order_frequency_flag" column
ords_prods_merge.head()
```
