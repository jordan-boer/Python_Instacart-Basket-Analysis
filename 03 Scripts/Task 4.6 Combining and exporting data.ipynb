## 04. Merging Data

```python
# Testing inner join merge with key "product_id"
pd.merge(df_ords_prods, df_prods, on = 'product_id', indicator = True)
```

##### Test merge was successful. Now will merge both datatframes into a new dataframe.

```python
# Merging df_prods with df_ords_prods using key "product_id"
ords_prods_merge = df_ords_prods.merge(df_prods, on = 'product_id', indicator = True)
```

```python
# Checking new merged dataframe ords_prods_merge
ords_prods_merge.head()
```

```python
# Checking shape of ords_prods_merge
ords_prods_merge.shape
```

```python
# Confirming results of merge by checking "_merge" frequency
ords_prods_merge['_merge'].value_counts()
```

```python
# Creating outer join merge with key "product_id" to test "_merge" frequency
ords_prods_merge_outer = df_ords_prods.merge(df_prods, on = 'product_id', how = 'outer', indicator = True)
```

```python
# Comparing results of inner and outer merge by checking outer "_merge" frequency
ords_prods_merge_outer['_merge'].value_counts()
```

##### As we can see from comparing the inner and outer merges frequencies, there is not a full match in the inner merge because we can see there are 30,200 counts of left-only and 11 counts of right-only in the outer merge. So, those were left out of the inner merge.

```python
# Checking for missing values in ords_prods_merge_outer
ords_prods_merge_outer.isnull().sum()
```
