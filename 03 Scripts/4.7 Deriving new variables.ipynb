## 05. If-Statements with For-Loops

```python
# Finding frequency of "orders_day_of_week" column
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creating a list for the busiest day of the week
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Combining result list with ords_prods_merge dataframe
ords_prods_merge['busiest_day'] = result
```

```python
# Checking frequency of new column "busiest_day"
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

## Creating a "Busiest Days" Column

```python
# Creating a list for the 2 busiest days and 2 slowest days
result_1 = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value in [0,1]:
    result_1.append("Busiest days")
  elif value in [3,4]:
    result_1.append("Slowest days")
  else:
    result_1.append("Regularly busy")
```

```python
result_1
```

```python
# Combining result_1 list with ords_prods_merge dataframe
ords_prods_merge['busiest_days'] = result_1
```

```python
# Checking frequency of new column "busiest_days"
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

##### Using the frequency check of the "orders_day_of_week" column from the exercise I determined the the 2 busiest days are 0 and 1 (Saturday and Sunday, respectively) and the slowest days are 3 and 4 (Tuesday and Wednesday, respectively). After adding the 1 into the condition for the "busiest days" and the 3 into the condition for the "slowest days", the frequency shows the increase in the "busiest" and "slowest" labels and the decrease in the "regularly busy" label, as expected.

## Creating a "Busiest Period of Day" Column

```python
# Finding frequency of "order_hour_of_day" column
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating a list for the busiest periods of the day
result_hours = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    result_hours.append("Most orders")
  elif value in [23,6,0,1,5,2,4,3]:
    result_hours.append("Fewest orders")
  else:
    result_hours.append("Average orders")
```

##### Since we are using 3 labels, I broke the day down into 3 equal parts of 8 hours each. So I checked the frequency table for the top 8 hours with the highest frequency, which happens to be from 9am(9) - 4pm(16) and the 8 lowest frequency hours, 11pm(23) - 6am(6). Then, I used those values to set the conditions for making my list.

```python
result_hours
```

```python
# Combining result_hours list with ords_prods_merge dataframe
ords_prods_merge['busiest_period_of_day'] = result_hours
```

```python
# Checking frequency of new column "busiest_period_of_day"
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
