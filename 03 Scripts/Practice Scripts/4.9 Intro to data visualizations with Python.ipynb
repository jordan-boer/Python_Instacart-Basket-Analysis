## 03. Creating Bar Charts

```python
# Creating a bar chart for "orders_day_of_week" column
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Reordering bar chart by index order
ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Exporting bar chart and saving it as an image in projects folder
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

## 05. Sampling Data

```python
# Creating a list holding True/False values to the test np.random.rand() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the dataframe "big"
big = ords_prods_merge[dev]
```

```python
# Store 30% of the sample the dataframe "small"
small = ords_prods_merge[~dev]
```

```python
# Checking length of original dataframe
len(ords_prods_merge)
```

```python
# Checking that length of sample dataframes match original
len(big) + len(small)
```

```python
# Reducing "small" sample dataframe to only columns needed for chart.
df = small[['orders_day_of_week', 'prices']]
```

```python
# Checking reduced "small" dataframe
df.head()
```

## 06. Creating a Line Chart from Sample Dataframe

```python
# Creating a line chart from the "df" dataframe created above
line = sns.lineplot(data = df, x = 'orders_day_of_week', y = 'prices')
```
