# 07. Subsetting

```python
# Shows how Python looks for data returning "true" or "false" based on criteria
df_prods['department_id']==19
```

```python
# Enlcosing above command in another instance of "df_prods[]" tells Python to map true/false procedure onto columns within df_prods
# Results in a list of only true values
df_prods[df_prods['department_id']==19]
```

```python
# Assign results to new dataframe called df_snacks
df_snacks = df_prods[df_prods['department_id']==19]
```

```python
df_snacks.head()
```

```python
# Alternate subsetting using "loc"
df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19]
```

```python
# Alternate subsetting using "isin([])" - good for searching mulitple values
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
```
