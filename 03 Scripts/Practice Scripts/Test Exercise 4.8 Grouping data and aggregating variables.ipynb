## 04. Group By

```python
# Grouping df by column "product_name"
df.groupby('product_name')
```

## 05. Performing Single Aggregation

```python
# Grouping data by "department_id" and using "agg()" to obtain mean values of "order_number" column
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Same results with "mean()" function
df.groupby('department_id')['order_number'].mean()
```

## 06. Performing Multiple Aggregations

```python
# Multiple aggragations of "mean', "min", and "max"
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## 07. Aggregating Data with transform()

```python
# Grouping "user_id" column, applying transform() on "orders_number" column to get max orders, and creating new column "max_order"
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## 08. Deriving Columns with loc()

```python
# Creating "Loyal customer" condition
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Creating "Regular customer" condition
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Creating "New customer" condition
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
